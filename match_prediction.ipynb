```python
def rolling_averages(group, col, new_col):
    group = group.sort_values("date")
    #closed = 'left' assignes the value in next col
    rolling_stats = group[col].rolling(3, closed ='left').mean()
    group[new_col] = rolling_stats
    group = group.dropna(subset=new_col)
    return group
```

```python
col = ["gf","ga","sh","sot","dist","fk","pk","pkatt"]
new_col = [f"{c}_rolling" for c in col]
```

```python
new_col
```

```python
rolling_averages(group, col, new_col)
```

```python
matches_rolling = match_data.groupby("team").apply(lambda x: rolling_averages(x, col, new_col))
```

```python
matches_rolling = matches_rolling.droplevel("team")
```

```python
 matches_rolling.index = range(matches_rolling.shape[0])
```

```python
matches_rolling
```

```python
matches_rolling.dtypes
```

```python
def make_predictions(data, predictors):
    train = data[data["date"] < '2024-01-01']
    test = data[data["date"] > '2024-01-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted = preds), index = test.index)
    precision = precision_score(test["target"], preds)
    return combined, precision
    
```

```python
combined_col = predictors + new_col
combined, precision = make_predictions(matches_rolling, combined_col)
```
